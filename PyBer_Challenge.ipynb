### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import os
import matplotlib.pyplot as plt
import pandas as pd
import dataframe_image as dfi  # to save DataFrames as image files

# File to Load (Remember to change these)
city_data_to_load = os.path.join('.','resources','city_data.csv')
ride_data_to_load = os.path.join('.','resources','ride_data.csv')

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_by_type = pyber_data_df.groupby(['type'])['ride_id'].count()
# pyber_data_df['city'].value_counts()

rides_by_type
```

```python
# 2. Get the total drivers for each city type
drivers_by_type = city_data_df.groupby(['type'])['driver_count'].sum()

drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
fares_by_type = pyber_data_df.groupby(['type'])['fare'].sum()

fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
fares_per_ride_by_type = fares_by_type / rides_by_type

fares_per_ride_by_type
```

```python
# 5. Get the average fare per driver for each city type. 
fares_per_driver_by_type = fares_by_type / drivers_by_type

fares_per_driver_by_type
```

```python
#  6. Create a PyBer summary DataFrame.
pyber_summary_df = pd.DataFrame({
    'Total Rides':rides_by_type,
    'Total Drivers':drivers_by_type,
    'Total Fares':fares_by_type,
    'Average Fare per Ride':fares_per_ride_by_type,
    'Average Fare per Driver':fares_per_driver_by_type
})

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None

pyber_summary_df
```

```python
#  8. Format the columns.
the_columns = list(pyber_summary_df.columns)

for col in the_columns[0:2]:
    pyber_summary_df[col] = pyber_summary_df[col].map("{:,}".format)
for col in the_columns[2:5]:
    pyber_summary_df[col] = pyber_summary_df[col].map("${:,.2f}".format)

dfi.export(pyber_summary_df,os.path.join('.','analysis','pyber_summary.png'))

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 0. Read the merged DataFrame
pyber_data_df
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fares_by_type_and_date = pyber_data_df.groupby(['type','date'])['fare'].sum()

fares_by_type_and_date
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_by_type_and_date = fares_by_type_and_date.reset_index()

fares_by_type_and_date
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fares_by_type_and_date_pivot = fares_by_type_and_date.pivot(index='date', columns='type', values='fare')

fares_by_type_and_date_pivot.head(10)
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fares_by_type_and_date_pivot_datelim = fares_by_type_and_date_pivot.loc['2019-01-01':'2019-04-29',:]

fares_by_type_and_date_pivot_datelim
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fares_by_type_and_date_pivot_datelim.index = pd.to_datetime(fares_by_type_and_date_pivot_datelim.index)

fares_by_type_and_date_pivot_datelim
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
fares_by_type_and_date_pivot_datelim.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fares_per_week_by_type = fares_by_type_and_date_pivot_datelim.resample('W').sum()

fares_per_week_by_type.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')    # But what does that mean? What does that style do?

import datetime
# See https://stackoverflow.com/questions/57126647/how-to-change-xticks-to-yearly-interval-in-pandas-time-series-plot

fig, ax = plt.subplots(figsize=(18,6))

ax.set_title('Total Fare by City Type')
ax.set_ylabel('Fare($USD)')

ax.plot(
    fares_per_week_by_type
)

ax.legend(
    fares_per_week_by_type.columns, # List (index) of the labels for the legend
    title='type',
#     title_fontsize=12,
    loc='best'
)

x_ticks = pd.date_range(datetime.datetime(2019,1,1), datetime.datetime(2019,5,1), freq='MS')

ax.set_xticks(x_ticks.to_pydatetime())
ax.set_xticklabels([tk.strftime('%b') for tk in x_ticks])
ax.set_xlabel(x_ticks[0].strftime('%Y'), loc='left')

plt.savefig(
    os.path.join('.','analysis','total_fare_by_city_type.ax.png')
    , bbox_inches='tight'   # Prevents the saved image from cropping the text label
#     , pad_inches=0.25       # Adds a quarter-inch pad space around the edge of the figure
)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')    # But what does that mean? What does that style do?

fares_per_week_by_type.plot(
    figsize=(18,6),
    title='Total Fare by City Type',
    legend=True,
    xlabel='',
    ylabel='Fare($USD)'
)

plt.savefig(
    os.path.join('.','analysis','total_fare_by_city_type.png')
    , bbox_inches='tight'   # Prevents the saved image from cropping the text label
#     , pad_inches=0.25       # Adds a quarter-inch pad space around the edge of the figure
)
```
